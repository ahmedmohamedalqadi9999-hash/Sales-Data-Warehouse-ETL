```python
df_dim_product['product_sk']=None
df_dim_product['product_Bk']=df_temp['ProductID']
df_dim_product['product_Name']=df_temp['ProductName'].str.strip()
df_dim_product['Subcategory_Bk']=df_temp['SubCategoryID'].iloc[:, 0].values
df_dim_product['price']=df_temp['Price'] .astype('float') .fillna(0)
df_dim_product['Subcategory_Name']=df_temp['SubCategoryName'] 
df_dim_product['Category_Bk']=df_temp['CategoryID'].iloc[:, 0].values
df_dim_product['Category_Name']=df_temp['CategoryName']
```
